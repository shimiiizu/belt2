```python
import pandas as pd
import seaborn as sns
import cv2
import matplotlib.pyplot as plt

# 事前にcsvの保存方式をカンマ区切りに変える必要があるので注意
df = pd.read_csv('./No2_10_height_s.csv', encoding='shift_jis',delimiter=',',skiprows=51)
im = cv2.imread('./No2_10_color.JPG')
im_gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
```

```python
# リスト作成

list = []
for i in range(1024):
  list.append(i)
```

```python
df.columns = list
```

```python
sns.heatmap(df, cmap='hot')
```

```python
sns.heatmap(im_gray,cmap='Greys')
```

```python
fig, ax = plt.subplots()

x=im_gray.reshape(1,im_gray.size)
y=df.values.reshape(1,df.values.size)

ax.scatter(x, y)
ax.set_xlabel('gray')
ax.set_ylabel('height')
plt.show()
```
